```python
#Importando as bibliotecas que serão utilizadas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df_gpd = pd.read_csv('/content/drive/MyDrive/CIENTISTA DE DADOS/PROJETOS/PIB/gdp_per_capita.csv', thousands=',', encoding='latin1',delimiter='\t')
```

```python
df_oecd = pd.read_csv('/content/drive/MyDrive/CIENTISTA DE DADOS/PROJETOS/PIB/oecd_bli_2015.csv', thousands=',')
```

```python
df_gpd.rename(columns={"2015":"GPD per capita"}, inplace=True)
df_gpd.set_index("Country", inplace=True)
```

```python
DFF = df_oecd.loc[df_oecd.INEQUALITY == 'TOT'].pivot(index="Country", columns="Indicator", values="Value")
DFF
```

```python
DFFF = pd.merge(left=DFF, right=df_gpd, left_index=True, right_index=True)
```

```python
DFFF
```

```python
remove_indices = [0, 1, 6, 8, 33, 34, 35]
keep_indices = list(set(range(36)) - set(remove_indices))

GG = DFFF[["GPD per capita", 'Life satisfaction']].iloc[keep_indices]
GG
```

```python
X = GG['GPD per capita'].values.reshape(-1,1)
y = GG['Life satisfaction'].values.reshape(-1,1)
```

```python
modelo_lm = LinearRegression()
```

```python
modelo_lm.fit(X,y)
```

```python
predicao = [[22587]]
print(modelo_lm.predict(predicao))
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
modelo_knn = KNeighborsRegressor(n_neighbors=3)
```

```python
modelo_knn.fit(X,y)
```

```python
modelo_knn.predict(predicao)
```
